```python
import random
import time
#from sigfig import round

st = time.time() #start timer

#T = 298.15 #K #range: 298.15 - 318.15

Tmin = 293.15
Tmax = 318.15
P = 1.01325 #Bar

m = 0 #mixture counter
nmin = 3

#no. of z samples
num_samples_base2 = 8
num_samples = 2**num_samples_base2 ###

print('no. of z samples:',num_samples) #number of global component, z samples
print('no. of total samples:',num_samples*binary_mix_no)

master_list = []
while m <= (binary_mix_no-1): #goes through each binary mixture
    #print(binary_mix_list[m])
    mix_func_no_ib=binary_mix_list[m]
    #mix_func = add_ibuprofen(binary_mix_list[m])
    #print(mix_func.nc)
    mix_func_no_ib.original_unifac()
    eos = virialgamma(mix_func_no_ib, virialmodel='ideal_gas', actmodel='original_unifac') #choosing activity coef model to produce data
    
    z_counter = 0 # reset z counter
    
    #random z generator
    sampler_sob = qmc.Sobol(d=4, scramble=True, seed = m)
    sample_sob = sampler_sob.random_base2(num_samples_base2)
    final_sample_list = []
    for sample in sample_sob:
        x1 = sample[0]/(sample[0] + sample[1] + sample[2])
        x2 = sample[1]/(sample[0] + sample[1] + sample[2])
        x3 = sample[2]/(sample[0] + sample[1] + sample[2])
        Tsob = (sample[3] * (Tmax-Tmin) + Tmin)
        final_sample_list.append([x1, x2, Tsob])


    #display samples taken for each mixture
#     plt.figure(m)
#     plt.scatter(z_sob, T_sob,marker='x',s=15)
    
    
    while z_counter <= num_samples-1:
        #z = np.array(z_rand_list[z_counter])
        z1 = final_sample_list[z_counter][0]
        z2 = final_sample_list[z_counter][1]
        #z3 = final_sample_list[z_counter][2]
        #z = np.array([z1,z2,z3])
        z = np.array([(z1/(z1 + z2)),(z2/(z1 + z2))])
        T = final_sample_list[z_counter][2]
        #T = 300
        tpd_m = tpd_minimas(nmin, z, T, P, eos, stateW='L', stateZ='L')
        tpd_values = tpd_m[-1] #outputs nmin TPD values 
        min_tpd_value = min(tpd_values) #takes the minimum TPD value
        if min_tpd_value < -1e-8:
            ismisc = 0
        else:
            ismisc = 1
        min_tpd_value_index = np.where(tpd_values == min_tpd_value) #indexes the minimum TPD value, so its corresponding w can be displayed      
        min_tpd_value_index = min_tpd_value_index[0][0] #for repeat solutions, only takes first in the list
        tpd_m_w = tpd_m[0][min_tpd_value_index]
        tpd_m_w = tpd_m_w.tolist() #converts type from array to list
        
        #master_list.append([T,mix_func_no_ib.names[0],mix_func_no_ib.names[1],z1,z2,min_tpd_value,ismisc]) #takes only lowest tpd;might miss result if 2 minimum TPD values are identical, but have different w compositions
        master_list.append([T,mix_func_no_ib.names[0],mix_func_no_ib.names[1],z1,z2,min_tpd_value,ismisc]) #only shows first two minima solutions
        z_counter = z_counter+1
    m=m+1
    #print()

et = time.time()
elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds') #execution time
```

```python
import random
import time
#from sigfig import round

st = time.time() #start timer

#T = 298.15 #K #range: 298.15 - 318.15

Tmin = 293.15
Tmax = 318.15
P = 1.01325 #Bar

m = 0 #mixture counter
nmin = 3

#no. of z samples
num_samples_base2 = 7
num_samples = 2**num_samples_base2 ###

print('no. of z samples:',num_samples) #number of global component, z samples
print('no. of total samples:',num_samples*ternary_mix_no)

master_list = []
while m <= (ternary_mix_no-1): #goes through each ternary mixture
    mix_func = ternary_mix_list[m]
    #mix_func = add_ibuprofen(ternary_mix_list[m])
    mix_func.original_unifac()
    eos = virialgamma(mix_func, virialmodel='ideal_gas', actmodel='original_unifac') #choosing activity coef model to produce data
    
    z_counter = 0 # reset z counter
    
    #random z generator
    sampler_sob = qmc.Sobol(d=4, scramble=True, seed = m)
    sample_sob = sampler_sob.random_base2(num_samples_base2)
    final_sample_list = []
    for sample in sample_sob:
        x1 = sample[0]/(sample[0] + sample[1] + sample[2] + sample[3])
        x2 = sample[1]/(sample[0] + sample[1] + sample[2] + sample[3])
        x3 = sample[2]/(sample[0] + sample[1] + sample[2] + sample[3])
        x4 = sample[3]/(sample[0] + sample[1] + sample[2] + sample[3])
        final_sample_list.append([x1, x2, x3])


    #display samples taken for each mixture
#     plt.figure(m)
#     plt.scatter(z_sob, T_sob,marker='x',s=15)
    
    
    while z_counter <= num_samples-1:
        #z = np.array(z_rand_list[z_counter])
        z1 = final_sample_list[z_counter][0]
        z2 = final_sample_list[z_counter][1]
        z3 = final_sample_list[z_counter][2]
        #z4 = final_sample_list[z_counter][3]
        #z = np.array([z1,z2,z3,z4])
        z = np.array([(z1/(z1+z2+z3)),(z2/(z1+z2+z3)),(z3/(z1+z2+z3))])
        #T = T_sob[z_counter]
        T = 300
        tpd_m = tpd_minimas(nmin, z, T, P, eos, stateW='L', stateZ='L')
        tpd_values = tpd_m[-1] #outputs nmin TPD values 
        min_tpd_value = min(tpd_values) #takes the minimum TPD value
        if min_tpd_value < 0:
            ismisc = 0
        else:
            ismisc = 1
        min_tpd_value_index = np.where(tpd_values == min_tpd_value) #indexes the minimum TPD value, so its corresponding w can be displayed      
        min_tpd_value_index = min_tpd_value_index[0][0] #for repeat solutions, only takes first in the list
        tpd_m_w = tpd_m[0][min_tpd_value_index]
        tpd_m_w = tpd_m_w.tolist() #converts type from array to list
        
        master_list.append([T,mix_func.names[0],mix_func.names[1],mix_func.names[2],z1,z2,z3,min_tpd_value,ismisc]) #takes only lowest tpd;might miss result if 2 minimum TPD values are identical, but have different w compositions
        #~master_list.append([T,mix_func.names,[z[0],z[1]],tpd_m[0][0],tpd_m[0][1],tpd_values[0],tpd_values[1]]) #only shows first two minima solutions
        z_counter = z_counter+1
    m=m+1
    #print()

et = time.time()
elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds') #execution time
```
